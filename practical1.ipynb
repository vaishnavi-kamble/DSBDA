```python
df['Paid'] = df['Paid'].fillna(df['Paid'].mean())
df['like'] = df['like'].fillna(df['like'].mean())
df['share'] = df['share'].fillna(df['share'].mean())
df.isnull().sum()
```

```python
subset_2 = df[['Page total likes','Post Weekday', 'Lifetime Post Consumers']]
subset_2
```

```python
subset_3 = df[(df['Type'] == 'Status') & (df['Post Hour'] == 10)]
subset_3
```

```python
subset_4 = df[(df['comment'] > 30) &  (df['share'] > 250)]
subset_4
```

```python
subset_5 = df[(df['Type'] == 'Photo') & (df['Lifetime Post Consumers'] > 1000)]
subset_5
```
