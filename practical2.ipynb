```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("pima-indians-diabetes.csv",names=['Number of times pregnant','Plasma glucose concentration a 2 hours in oral glucose tolerance test','Diastolic blood pressure (mm Hg)',''','2-Hour serum insulin (mu U/ml) ','Body mass index (weight in kg/(height in m)^2)','Diabetes pedigree function','Age (years)','Class variable (0 or 1)'])
data
```

a.Data Cleaning

```python
data.isna().sum()
```

```python
data.info()
```

```python
data.shape
data
```

```python
data.describe()
```

```python
#removing first 9 rows and resseting the index
data = data.iloc[9:,:] 
data =data.reset_index(drop=True)
data
```

```python
#Fills missing values with the mean of each column.
data = data.fillna(value = data.mean(numeric_only = True))
data
```

```python
data.isnull().sum()
```

b.Data Integration

```python
subs1 = data[data['Age (years)']>50]
#subs1.shape
```

```python
subs2 = data[data['Diastolic blood pressure (mm Hg)']>60]
#subs2.shape
```

```python
merged = pd.concat([subs1,subs2])
merged
```

```python
#correlation matrix , Visualizes correlations between variables.
corr_mat = sns.heatmap(data=data.corr(),annot=True)
corr_mat
```

```python
#Splitting Data for Model Training
from sklearn.model_selection import train_test_split
x=data.iloc[:,0:8]
y = data['Class variable (0 or 1)']
train_x,test_x,train_y,test_y = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
train_x
```

```python
train_y
```

```python
#Creates and trains a Decision Tree Classifier.
from sklearn import tree
model = tree.DecisionTreeClassifier()
model.fit(train_x,train_y)
```

```python
#Predicts diabetes (0 or 1) for test data.
ypred = model.predict(test_x)
```

```python
#Evaluating Model Performance
from sklearn.metrics import confusion_matrix,accuracy_score,recall_score,precision_score,f1_score
matrix = confusion_matrix(test_y,ypred)
matrix
```

```python
accuracy_score(test_y,ypred)
```

```python
recall_score(test_y,ypred)
```

```python
precision_score(test_y,ypred)
```

```python
f1_score(test_y,ypred)
```

```python
#Displays the importance of each feature in decision making.
model.feature_importances_
```

```python
from sklearn.tree import plot_tree
plt.figure(figsize=(20,16))
plot_tree(model, filled=True, feature_names=x.columns, class_names=['No Diabetes', 'Diabetes'])
plt.show()
```
