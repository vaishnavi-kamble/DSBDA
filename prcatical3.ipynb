```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('cancer_data_set.csv')
df.head()
```

```python
df.dtypes
```

```python
#Remove missing values
df_cleaned = df.dropna()
df_cleaned
```

```python
# Count Plot for Diagnosis Distribution
plt.figure(figsize=(6, 4))
sns.countplot(x=df["diagnosis"], palette="coolwarm")
plt.title("Diagnosis Count")
plt.xlabel("Diagnosis (B = Benign, M = Malignant)")
plt.ylabel("Count")
plt.show()


#The count plot shows that benign tumors are more frequent than malignant ones. 
#helping to check class balance in the dataset.
```

```python
#Pair plot for selected features
selected_features = ['radius_mean', 'texture_mean', 'perimeter_mean', 'area_mean', 'smoothness_mean']
sns.pairplot(df_cleaned, vars=selected_features, hue='diagnosis', palette={'B': 'blue', 'M': 'red'}, diag_kind='hist')
plt.show()

#This pair plot visualizes relationships between selected features, showing how benign (B) and malignant (M) cases are distributed,
#helping identify patterns and separability
```

```python
#Correlation Heatmap -> This shows how features are related to each other.

plt.figure(figsize=(12, 8))

# Convert diagnosis column ('M' -> 1, 'B' -> 0)
df_numeric = df.copy()
df_numeric["diagnosis"] = df_numeric["diagnosis"].map({"M": 1, "B": 0})

# Compute correlation matrix without 'id'
corr_matrix = df_numeric.drop(columns=["id"]).corr()

# Plot heatmap
sns.heatmap(corr_matrix, cmap="coolwarm", annot=False)
plt.title("Feature Correlation Heatmap")
plt.show()



#The correlation heatmap confirms that features like radius_mean, perimeter_mean, and area_mean are highly correlated.
```

```python
sns.pairplot(df[['radius_mean', 'texture_mean', 'area_mean', 'smoothness_mean', 'diagnosis']], hue="diagnosis", palette="coolwarm")
plt.show()

#This pair plot shows relationships between selected features, helping visualize how benign (0) and malignant (1) cases are distributed and whether features can separate the two classes.
# This helps visualize how well different features separate benign and malignant cases
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(x=df['id'].astype(str), y=df['concavity_mean'], color='blue', label='Concavity Mean')
sns.barplot(x=df['id'].astype(str), y=df['concavity_worst'], color='red', alpha=0.6, label='Concavity Worst')

plt.xlabel("ID")
plt.ylabel("Concavity Values")
plt.title("Concavity Mean vs. Concavity Worst")
plt.legend()

plt.show()

#This bar plot compares concavity mean (blue) and concavity worst (red) across different patient IDs, helping visualize variations in concavity measurements for each case.
```

```python
#Histogram of Radius Mean → Displays the frequency distribution of radius mean.
plt.figure(figsize=(6, 5))
sns.histplot(df['radius_mean'], bins=30, kde=True, color='purple')
plt.title("Distribution of Radius Mean")
plt.xlabel("Radius Mean")
plt.ylabel("Frequency")
plt.show()

#The histogram indicates that the majority of tumors have a radius mean between 10 and 20.
# highlighting its frequency across different value
```

```python
#Box Plot of Area Mean by Diagnosis
#This helps compare area_mean values for Malignant vs. Benign cases.\

plt.figure(figsize=(6, 4))
sns.boxplot(x=df["diagnosis"], y=df["area_mean"], palette="coolwarm")
plt.title("Area Mean by Diagnosis")
plt.xlabel("Diagnosis")
plt.ylabel("Area Mean")
plt.show()

# The box plot suggests that malignant tumors tend to have a higher area_mean compared to benign ones.
#making it a key distinguishing feature
```

```python
plt.figure(figsize=(6,5))
sns.violinplot(x='diagnosis', y='compactness_mean', data=df, palette="muted")
plt.title("Compactness Mean vs Diagnosis")
plt.xlabel("Diagnosis (0 = Benign, 1 = Malignant)")
plt.ylabel("Compactness Mean")
plt.show()

#This violin plot visualizes the distribution of compactness mean for benign (0) and malignant (1) cases, showing density, spread, and potential differences between the two classes.
```

```python
#Scatter Plot 

plt.figure(figsize=(6,5))
sns.scatterplot(x='radius_mean', y='perimeter_mean', hue='diagnosis', data=df, palette='coolwarm')
plt.title("Radius Mean vs Perimeter Mean")
plt.xlabel("Radius Mean")
plt.ylabel("Perimeter Mean")
plt.show()

#This scatter plot shows the relationship between radius mean and perimeter mean,
#highlighting how benign (0) and malignant (1) cases are distributed and whether they form distinct clusters.
```

```python
plt.figure(figsize=(6,5))
sns.swarmplot(x='diagnosis', y='concave points_mean', data=df, palette="coolwarm")
plt.title("Concave Points Mean vs Diagnosis")
plt.xlabel("Diagnosis (0 = Benign, 1 = Malignant)")
plt.ylabel("Concave Points Mean")
plt.show()

# Malignant tumors (red) have higher concave points mean, showing a taller distribution, indicating more irregular tumor boundaries. Benign tumors (blue) have lower concave points mean, forming a shorter and wider cluster near the bottom, suggesting smoother edges.
```
